# Correlation
Correlations can be done on two continuous variables, to determine the relationship between them. As a reminder, a correlation can be between zero and one, and either positive or negative. The larger it is, the more closely related the two variables are.

```python
# Run correlation between number of cabins and passengers

cruise['Cabins'].corr(cruise['passngrs'])
```

### Use a correlation matrix to test all continuous data

```python
# Drop Categorical Data
cruise1 = cruise.drop(['Ship','Line'], axis = 1)
```

```python
# Create correlation using pearsons method
cruise1.corr(method='pearson')
```

```python
# Make the Matrix easier to read
cruise1.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

```python
# Correlation matrix using seaborn package
sns.heatmap(cruise1.corr(), annot=True)
```
