## Negative Skew

```python
# Startin with YearBlt
cruise.YearBlt.hist()
```

```python
# Squre YearBlt to transform the variable
# Squre the variable
cruise['YearBltSQ'] = cruise['YearBlt']**2
```

```python
cruise.YearBltSQ.hist()
```

Still about the same negative skew

```python
# Cube the data to transfrom the negative skew
cruise['YearBltCUBE'] = cruise['YearBlt']**3
```

```python
cruise.YearBltCUBE.hist()
```

```python
# BoxCox Transformation of Negative Skew
cruise['YearBltLog1'] = boxcox(cruise['YearBlt'],0)
```

```python
cruise.YearBltLog1.hist()
```

### None of the transformation methods really cented the YearBlt Data very well, use this data with caution.
