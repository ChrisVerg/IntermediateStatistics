#### Import Data and Packages

```python
import pandas as pd
import seaborn as sns
import numpy as np
# For BoxCox 
from scipy import stats
from scipy.stats import boxcox
```

```python
parks = pd.read_csv('C:/Users/chris/Desktop/data/DS105/Lesson2/Seattle_ParksnRec.csv')
```

```python
parks.head()
```

# Winter Trips

```python
# Assess the Skew of '# of trips Winter'

parks['# of trips Winter'].hist()
```

'# of trips Winter' appears to have bimodal distribution or and Positive Skew

## Squre Root Transformation

```python
# Square Root to Transform the data
parks['WinterTripsSQRT'] = np.sqrt(parks['# of trips Winter'])
```

```python
# View Squared Winter Data
parks.WinterTrips.hist()
```

```python
sns.distplot(parks['WinterTripsSQRT'])
```

## Logarithmic

```python
# Try a Logarithmic Transformation
parks['WinterTripsLog'] = np.log(parks['# of trips Winter'])
```

```python
parks.WinterTripsLog.hist()
```

Log creates Negative Skew

```python
sns.distplot(parks['WinterTripsLog'])
```

### The Logarithmic is close to normally distributed but the data is still bimodal for both transformations. 

# Winter Participants
